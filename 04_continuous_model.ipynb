
# Continuous Market Models

The tools to model a time continuous market model have now been presented. 
A *market model* will be characterized by an Ito diffusion process $X_t$ 
and by another stochastic process $B_t$:

$$
\begin{eqnarray*}
    dB_t &=& r_t B_t dt \\
    \frac{dS_t}{S_t} &=& \mu (t) dt + \sigma(t)dW_t
\end{eqnarray*}
$$

The stochastic process $S_t$ can model equity assets, fixe-income rates, 
foreign exchange currencies or other unobservable random variable (such
as a stochastic volatiliy). $B_t$ is the *money market account* which
when you invest $1$ at time $t$ gives $1+r_tdt$ at $t+dt$, with $r_t$
the instantaneous interest rate at $t$. We define the **discount factor**
from $0$ to $t$ by:

$$D_{0,t} = \frac{1}{B_t} = e^{-\int_0^t r_s ds}$$

The fact that the equation for $dB_t$ does not have a Brownian Motion term
means that $B_t$ is a non-risky asset with a fixed return $r_t$. However
$r_t$ is not necessarily a deterministic function and may depend on the 
risky asset (the instantaneous rate $r_t$ can therefore be a stochastic
process).

## Self financed portfolio

A portfolio strategy is self is a strategy that can buy or sell the risky asset,
lend or borrow money from the bank, but whose value is not modified by addition
or withdrawal of cash.

$V_t$ denotes the value or mark-to-market of this portfolio at time $t$.
$\delta_t$ is the number of stock we hold until the next decision date.

On a very short time scale, the variation of the value of a self-financed portfolio
is impacted by the move of the stock price, and interest paid by the bank on our
deposit. So:
$$dV_t = \delta_t dS_t + (V_t-\delta_t S_t)r dt$$

An **arbitrage** is a self-financed portfolio which verifies:

$$
\begin{eqnarray*}
  V_0 &=& 0 \\
  \mathbb{P}(V_T > 0) &>& 0 \\
  \mathbb{P}(V_T \leq 0) &=& 0
\end{eqnarray*}
$$

## Replication EDP

We saw that pricing is closely linked to the vaue of a replication portfolio
(perfect if possible). Mathematically, the problem is to find a self-financed 
replicating porfolio that perfectly replicate final value of the derivative.

If $\Phi$ is the payoff function of the derivative, we want:

$$
\left\{
 \begin{array}{ll}
   dV_t(t,S_t) &=& \delta_t dS_t + (V(t,S_t) -\delta_t S_t) r dt  \\
   V(T,S_T) &=& \Phi(S_T)
\end{array}
\right.
$$

The existence of such a solution is not intuitive. But this unicity is a 
consequence of AoA. Indeed, if two replicating porfolios exist, their
difference is still a self-financed portolio, whose value at $T$ equals
$0$ in every possible scenario. If their values are not the same at all
time, it's easy to build an arbitrage by selling the more expensive and
buying the cheapest.

### Resolution

Let's apply Ito on $V(t,S_t)$:

$$
\begin{array}{ll}
dV(t,S_t) &=& V_t'dt + V_x' dS_t +  \frac{1}{2}V_{xx}''(dS_t)^2 \\
dV(t,S_t) &=& \bigg(V_t'+ V_x'\mu S +  \frac{1}{2}V_{xx}''\sigma^2 S^2\bigg)dt + V_x'\sigma S dW_t
\end{array}
$$

But as $V$ is a self-financed portfolio, we know that:
$$
\begin{array}{ll}
dV_t(t,S_t) &=& \delta_t dS_t + (V(t,S_t) -\delta_t S_t) r dt \\
dV_t(t,S_t) &=& \bigg( \delta_t \mu S +  rV(t,S_t)-r\delta_t S_t \bigg) dt + \delta_t \sigma S dW_t 
\end{array}
$$

We have two stochastique differential equations for $V(t,S_t)$. As the 
$dW_t$ part can't be explained by a part of the form $dt$, i.e. there is
unicity of the decomposition into a brownian part and a deterministic 
part.

By comparing the $dW_t$ part we find that we necessarily have 
$V_x'(t,S_t) = \delta_t$.

By comparting the $dt$ parts, we should have:

$$
\begin{array}{rl}
V_t'+ V_x'\mu S +  \frac{1}{2}V_{xx}''\sigma^2 S^2 &=&\delta_t \mu S +  rV-r\delta_t S_t \\
V_t'+ V_x'\mu S +  \frac{1}{2}V_{xx}''\sigma^2 S^2 &=&V_x' \mu S +  rV-rV_x' S_t \\
V_t'+   \frac{1}{2}V_{xx}''\sigma^2 S^2 &=&  rV-rV_x' S_t \\
V_t'+ rV_x' S_t +  \frac{1}{2}V_{xx}''\sigma^2 S^2 - rV &=&  \\
V_t'+ rV_x' S_t +  \frac{1}{2}V_{xx}''\sigma^2 S^2 - rV &=& 0 \\
\end{array}
$$

## Risk Neutral probability

Let's have a look few seconds on the final property of this replication portfolio.

Suppose for a moment that the risky asset diffusion has a $r$ drift instead of
$\mu$. Call it $\hat{S}$ to avoid any mistake.

$$
\left\{
 \begin{array}{ll}
   \frac{d\hat{S}_t}{\hat{S}_t} &=& rdt + \sigma dW_t  \\
   \hat{S}_0 &=& S_0
\end{array}
\right.
$$

Apply Ito on the process $\hat{V}(t,\hat{S}_)=e^{r(T-t)}V(t,\hat{S}_t)$.

$$
\begin{array}{rl}
d\hat{V}(t,\hat{S}_t) &=& \big(-re^{r(T-t)}V_t +e^{r(T-t)}V_t'\big) dt + e^{r(T-t)}V_x' d\hat{S}_t + \frac{1}{2}e^{r(T-t)} V_{xx}''(d\hat{S}_t)^2 \\
~~ &=& e^{r(T-t)}\bigg(-rV_t + V_t' + V_x' r S + \frac{1}{2}V_{xx}''\sigma^2 S^2 \bigg) dt + e^{r(T-t)} V_x' \sigma dW_t \\
~~ &=& e^{r(T-t)} V_x' \sigma dW_t \\
\hat{V}(T,\hat{S}_T) -  \hat{V}(0,\hat{S}_0) &=& \int_0^T e^{r(T-t)} V_x' \sigma dW_t \\
\mathbb{E}\bigg[\hat{V}(T,\hat{S}_T) -  \hat{V}(0,\hat{S}_0)\bigg] &=& \mathbb{E}\bigg[ \int_0^T e^{r(T-t)} V_x' \sigma dW_t\bigg]=0 \\
\hat{V}(0,\hat{S}_0) &=&  \mathbb{E}\big(\hat{V}(T,\hat{S}_T)\big) = \mathbb{E}\big(e^{r(T-T)}\Phi(\hat{S}_T)\big)  = \mathbb{E}\big(\Phi(\hat{S}_T)\big) \\
e^{rT}V(0,S_0) &=&  \mathbb{E}\big(\Phi(\hat{S}_T)\big) \\
V(0,S_0) &=&  e^{-rT}\mathbb{E}\big(\Phi(\hat{S}_T)\big) \\
\end{array}
$$

The stochastic process $\hat{S}$ is a technical tool that allows us to express the pricing
as the expectation of the derivative payoff. Actually, this is possible to build a new 
probabiliy $\mathbb{Q}$ under wich the original process as a new dynamic:
$$\frac{dS_t}{S_t} = rdt + \sigma(t,S_t)dW_t$$

With $W$ a $\mathbb{Q}$-Brownian motion. We can reformulate the final equality:

$$
\Pi(0,S_0) =  e^{-rT}\mathbb{E}^{\mathbb{Q}}\big(\Phi(S_T)\big)
$$
